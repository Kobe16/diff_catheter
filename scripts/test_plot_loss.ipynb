```python
import matplotlib.pyplot as plt
import pickle
import os

from catheter_reconstruction.utils import *
```

```python
def plot_loss(loss_list, log_scale=False):
    """
    Plot the loss of each iteration
    """  
    
    ylabel = 'Error (mm)'
    title = '3D Tip Euclidean Distance Loss History'
            
    # Plot the loss
    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)
    fig1.suptitle(title)
    ax1.plot(loss_list)
    ax1.set_xlabel('Iteration')
    ax1.set_ylabel(ylabel)
    ax1.set_xlim([0, len(loss_list)])
    # ax1.set_ylim([0, 80])
    # ax1.set_ylim(bottom=0)
    if log_scale:
        # ax1.set_ylim(bottom=1)
        ax1.set_yscale('log')
        # ax1.set_ylabel(ylabel + ' (log scale)')
    ax1.grid(True)
    
    plt.show()
```
